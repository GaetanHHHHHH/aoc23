## Pandas

```python
inpute2 = re.split(r'(?=G)', inpute)
```

```python
inpute2.pop(0)
```

```python
for i in range(0, len(inpute2)):
    inpute2[i] = inpute2[i].replace('\n', "")
    inpute2[i] = re.split(': |, |; ', inpute2[i])
```

```python
columns = ['game', 'blue', 'red', 'green']
df = pd.DataFrame(columns=columns)
```

```python
for row in inpute2:
    game = row[0].split()[1]
    new_row = {'game': game, 'blue': 0, 'red': 0, 'green': 0}
    #print(counts)
    
    # Extract counts for each color from the row
    for entry in row[1:]:
        count_str, color = entry.split()
        if int(count_str) > new_row[color]:
            new_row[color] = int(count_str)

    max_index = df.index.max() if not df.empty else -1
    max_index += 1
    df.loc[max_index] = new_row
```

```python
df
```

```python
counts = 0

for index, row in df.iterrows():
    if (row['blue'] <= 14 and row['red'] <= 12 and row['green'] <= 13):
        counts += int(row['game'])
```

```python
print(counts)
```

```python
answer2 = 0

for index, row in df.iterrows():
    answer2 += row['blue']*row['red']*row['green']
```
